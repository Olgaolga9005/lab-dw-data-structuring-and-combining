```python
def clean_cols(url):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()
    df.rename(columns={"ST": "state"}, inplace=True)
    df.columns = df.columns.str.replace(" ", "_").str.lower()
    return df

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = clean_cols(url1)
df2 = clean_cols(url2)
df3 = clean_cols(url3)

df_new = pd.concat([df1, df2, df3], ignore_index=True)

df_new.to_csv("customer_cleaned.csv", index=False)

def clean_rows(df):

    df["gender"] = df["gender"].replace({
        "Femal": "F",
        "female": "F",
        "Male": "M"
    })

    cat_cols = ["customer", "state", "gender", "education", "policy_type", "vehicle_class"]
    df[cat_cols] = df[cat_cols].fillna("Unknown")

    df["customer_lifetime_value"] = (df["customer_lifetime_value"].str.replace("%", "", regex=False).astype(float))
    df["number_of_open_complaints"] = (df["number_of_open_complaints"].str.split("/").str[1].fillna(0).astype(int))
    
    num_cols = ["customer_lifetime_value","income","monthly_premium_auto","total_claim_amount","number_of_open_complaints"]
    df[num_cols] = df[num_cols].fillna(df[num_cols].mean()).round()
    
    df["state"] = df["state"].replace({"Cali": "California","AZ": "Arizona","WA": "Washington"})
    df["education"] = df["education"].replace({"Bachelors": "Bachelor"})
    df["vehicle_class"] = df["vehicle_class"].replace({"Luxury SUV": "Luxury","Sports Car": "Luxury","Luxury Car": "Luxury"})

    return df


df_new = clean_rows(df_new)
df_new.head()
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
rev_channel = pd.pivot_table(
    df,
    values="customer_lifetime_value",
    values="gender";
    index="sales_channel",
    aggfunc="sum"
).round(2)

total_revenue = rev_channel["customer_lifetime_value"].sum()

rev_channel["revenue_%"] = (rev_channel["customer_lifetime_value"] / total_revenue * 100).round()

rev_channel
#The Agent and Branch channels are the two most effective in terms of revenue. 
# The Agent channel generated 36% more revenue than the Branch channel.
```

```python
average = pd.pivot_table(
    df,
    values="customer_lifetime_value",
    index="gender",         # rows
    columns="education",    # columns
    aggfunc="mean"
).round(2)

average
```

```python
pivot_sum = pd.pivot_table(
    df,
    values="customer_lifetime_value",
    index="gender",
    columns="education",
    aggfunc="sum"
)
pivot_pct = (pivot_sum / pivot_sum.values.sum() * 100).round(2)

pivot_pct
```

The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*
